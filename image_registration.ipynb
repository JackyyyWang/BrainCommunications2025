```python
import os
import nibabel as nib
import numpy as np
import ants
```

```python
folder_list = sorted(os.listdir('Phase_Imaging_Project')) ## change folder name here
folder_list[1:]
```

```python

for i in folder_list[1:]:
    subject_name = sorted(os.listdir(i))[1].split('_')[0]
    y1_subject_name = sorted(os.listdir(i))[-1].split('-')[1]
    print(subject_name)
    print(y1_subject_name)
    swi_b = subject_name + '_SWI.nii'
    swi_y = 'Y1-' + y1_subject_name + '-SWI.nii'
    flair_y = 'Y1-' + y1_subject_name + '-FLAIR_reg_SWI.nii.gz'
    
    swi_b = os.path.join(i, swi_b)
    swi_y = os.path.join(i, swi_y)
    flair_y = os.path.join(i, flair_y)

    fixed = ants.image_read(swi_b)
    m_swi_y = ants.image_read(swi_y)
    m_flair_y = ants.image_read(flair_y)

    
    mytx_swi_y = ants.registration(fixed=fixed, moving=m_swi_y, type_of_transform='Rigid', verbose=True)
    mv_swi_y = mytx_swi_y['warpedmovout']
    ants.image_write(mv_swi_y, os.path.join(i, y1_subject_name+'_swi_y1_reg.nii'))

    mv_flair_y = ants.apply_transforms(fixed=fixed, moving=m_flair_y,
                                      transformlist=mytx_swi_y['fwdtransforms'])

    ants.image_write(mv_flair_y, os.path.join(i, y1_subject_name+'_flair_swi_y1_reg.nii'))

    
```
